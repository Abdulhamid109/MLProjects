Graphical representation (independent feature distribution)

```python
energy['latitude'].value_counts().plot(kind='barh', figsize=(8, 6))
plt.xlabel("Count", labelpad=14)
plt.ylabel("latitude", labelpad=14)
plt.title("Distribution of latitude", y=1.02)
```

```python
energy['longitude'].value_counts().plot(kind='barh', figsize=(8, 6))
plt.xlabel("Count", labelpad=14)
plt.ylabel("longitude", labelpad=14)
plt.title("Distribution of longitude", y=1.02)
```

**Insights**
here there is no missing values present in latitude longitude and usage 
minimum usage is around 0.3 KWhrs
maximum usage is around 522.1 KWhrs
latitude and longitude are equally distributed across the dataset
Hence data is not biased
